```python
publisher=best['publisher'].value_counts()
bars = plt.bar(publisher.index,publisher.values)
plt.figure(figsize=(25,15))
publisher.plot(kind='bar')
plt.title('출판사')
plt.ylabel('책출판수')
plt.xlabel('출판사')
plt.xticks(rotation=90)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{int(height)}',
             ha='center', va='bottom', fontsize=9)
plt.show()
```

```python
best['publish_month'] = pd.to_datetime(best['publishDate'], errors='coerce').dt.to_period('M')

plt.figure(figsize=(20, 10))
monthly_counts = best['publish_month'].astype(str).value_counts().sort_index()
bars = plt.bar(monthly_counts.index,monthly_counts.values)
monthly_counts.plot(kind='bar')
plt.title("월별 출간 도서 수")
plt.ylabel("도서 수")
plt.xlabel("년별 월")
plt.xticks(rotation=90)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{int(height)}',
             ha='center', va='bottom', fontsize=9)
plt.show()
```
